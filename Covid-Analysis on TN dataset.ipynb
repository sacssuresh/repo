```python
#Multiple Linear regression
X = complete_TN[['New cases', 'New recovered']]
y = complete_TN['Death']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
#Coefficient values indicate that a unit in death indicates that there is a decrease in new cases and increase in new recovered.
#However these values gives very little importance to death rate since they are under 1.
```

```python
#To make predictions
y_pred = regressor.predict(X_test)
```

```python
#Predict actual output of X_test with  y_pred
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
